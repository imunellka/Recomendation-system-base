### Модель

#### Apriori

##### Support

– Определяет популярность элемента по всем наборам.



##### Confidence

Вероятность того, что элемент 1 будет куплен после покупки элемента 0 { 0 -> 1}.

Измеряется, как кол-во появлений набор в покупках { яблоко - пиво } 3/4, т.е. 75%

![](https://annalyzin.files.wordpress.com/2016/03/association-rule-confidence-eqn.png?w=527&h=77)


##### Lift


Вероятность того, что элемент 1 будет куплен после покупки элемента 0 { 0 -> 1}, но учитывая вероятность товара 1.

Набор { яблоко -> пиво} = 1, т.к. это не очень популярный и маловероятный набор.

Если значение больше 1, то вероятность покупки больше, если ближе к 0, то меньше.


![](https://annalyzin.files.wordpress.com/2016/03/association-rule-lift-eqn.png?w=566&h=80)

```python
classic_df.head(10)
```

```python
# если сразу передать набор, то не получим результат
association_rules = classic_apriori(classic_df.head(10), min_support=0.3, min_confidence=0.2, min_lift=1, min_length=2)
association_results = list(association_rules)

for i in range(0, len(association_results)):
    print(association_results[i][0])
```

```python
# делаем подготовку данных
records = []
for i in range(0, 100):
    records.append([str(classic_df.values[i, j]) for j in range(0, 20)])


records[:3]
```

```python
# изменение min_support / min_confidece ведет к изменению размера релузьтата
association_rules = classic_apriori(records, min_support=0.04, min_confidence=0.2, min_lift=3, min_length=2)
association_results = list(association_rules)

print("Найдено {} правил".format(len(association_results)))

for i in range(0, len(association_results)):
    print(association_results[i][0])
```

```python
# получаем правила из общего результат и смотрим метрики
for item in association_results:
    pair = item[0]
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])

    print("Support: " + str(item[1]))


    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```

```python
# модель
# установим порог для support в 0.5
apriori(df,min_support=0.5)
```

```python
# такая же поддержка, но с использованием имени элементов
apriori(df,min_support=0.5, use_colnames=True)
```

```python
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```
